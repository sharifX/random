# Scientific Names tool  comparison 

A simple notebook showing how different tools provide different responses for species scientific names.

This is a test dataset (plants) that contain real names, wrong author names, typos. 
```
scientificName
Pinus mugo Turra
Pinus mugo Smith
Vicia faba L.
Vicia faba S.
Vicia faba
Vicia fava
```


````python
#GBIF API response 
from pygbif import species
import json

# This is a match and ACCEPTED name 
species.name_suggest(q='Pinus mugo Turra')
````

````python
# Try with different/wrong author name
# No match, no response 
species.name_suggest(q='Pinus mugo Smith')
````

````python
# Try with no author name 
result = species.name_suggest(q='Pinus mugo')
````

````python
for names in result:
    print(names['scientificName']+","+names['status'])
````

````python
# correct suffix, matches, ACCEPTED
species.name_suggest(q='Vicia faba L.')
````

````python
# try with a wrong suffix, no response
species.name_suggest(q='Vicia faba S.')
````

````python
# try with out any suffix 
faba = species.name_suggest(q='Vicia faba')
for names in faba:
    print(names['scientificName']+","+names['status'])
````

````python
#mistake
species.name_suggest(q='Vicia fava')
````
