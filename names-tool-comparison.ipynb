# Scientific Names tool  comparison <a id="Scientific-Names-tool--comparison"></a>

This notebook is part of ENVRI FAIR WP11 use case on scientific names and FAIR data. The goal is to document how different tools provide different responses for the same scientific name.

The following small test dataset (plants) were generated to highlight some of the issues. 

| scientificName                       | Notes                     |
|--------------------------------------|---------------------------|
| Pinus mugo Turra                     | Correct and accepted name |
| Pinus mugo                           | Exclude the author name   |
| Pinus mugo Smith                     | Wrong author name         |
| Pinus muco Turra                     |   Spelling mistake                         |
| Pino mugo Turra                      | Wrong genus                          |
| Pinus basso Turra        |              Wrong species             |

### Global Names Resolver <a href id="Global-Names-Resolver"></a>

*Pinus mugo Turra*

Accepted name with the binomials and author name (the person credited with the first formal use of the name)



````python
import requests
from pprint import pprint
import json
from requests.structures import CaseInsensitiveDict
url = "http://resolver.globalnames.org/name_resolvers.json?names=Pinus+mugo+Turra"
headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"

resp = requests.get(url, headers=headers)
# data_source_title shows the source name 
for i in json.loads(resp.text)['data']:
# print the first results
    pprint(i['results'][0])
    pprint(i['results'][1])
    pprint(i['results'][2])
    pprint(i['results'][3])
````

Now *Pinus mugo*

Excluding the author name, only the binomials.

````python
import requests
from pprint import pprint
import json
from requests.structures import CaseInsensitiveDict
url = "http://resolver.globalnames.org/name_resolvers.json?names=Pinus+mugo"
headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"

resp = requests.get(url, headers=headers)
# data_source_title shows the source name 
for i in json.loads(resp.text)['data']:
    pprint(i['results'])
   
````

### GBIF 

Using the python library 


````python
#GBIF API response 
from pygbif import species
import json

# This is a match and ACCEPTED name 
species.name_suggest(q='Pinus mugo Turra')
````


## A few other services to compare

#### NomenMatch 

A few other web friendly tools 
http://match.taibif.tw/about.html
"NomenMatch is a tool for taxonomists to match a set of species names against other sets with certain authority. It returns matched names, ids, usage status, and links to original sources. So not only the matched results, users can also know about differences of name usage among different sources. This is important because we believe that there is no absolute right or wrong, and only different perspectives in taxonomy. This tool provides easily references for taxonomists."

#### EU-NOMEN 

EU-NOMEN/PESI project 
http://www.eu-nomen.eu/portal/index.php

EU-nomen enables the correct use of species names and their classification, to more accurately manage information on animals and plants. This is the first all-taxa inventory for European species.


````python
import requests
from requests.structures import CaseInsensitiveDict
url = "http://www.eu-nomen.eu/portal/rest/PESIRecordsByName/Vicia%20Faba?like=true&offset=1"
headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"


resp = requests.get(url, headers=headers)
print(resp.status_code)
````

````python
json.loads(resp.text)
````

````python
for i in json.loads(resp.text):
    print(i['scientificname']+" "+i['valid_name']+" "+i['valid_authority'] )
````
